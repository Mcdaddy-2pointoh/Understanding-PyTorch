### 2. TENSOR Operations


There are two ways to multiply in matices, vectors or tensors:

* Element wise operation as done by `torch.mul(TENSOR, scalar)`


* Matrix multiplication [Dot Product] operation as done by `torch.matmul(TENSOR_A, TENSOR_B)`. This operation can be only done when shape of TENSOR_A is (m,n) and shape of TENSOR_B is (n,o) producing the resulting tensor (TENSOR_R) with a shape (m,o)

```python
# Mat multiplying a vector by its transpose using torch.transpose()
torch.matmul(TENSOR, torch.transpose(TENSOR, 0, 1))
```

```python
# Mat multiplying a vector by its transpose using TENSOR.T
torch.matmul(TENSOR, TENSOR.T)
```

```python
# Understanding the change in TENSOR due to transpose
print(TENSOR)
print(TENSOR.T)
```

```python
# Understanding the change in shape of a transpose
print(TENSOR.shape, TENSOR.T.shape)
```

## k. Reshape, Stack, Squeeze, Unsqueeze

* Reshape - Reshape an input tensor to a defined shape

* View - Return a view of a tensor but keep the same memory as original.

* Stacking - Combine multiple tensors horizontally or vertically. Combine on a dimension of the tensor.

* Squeeze - Removes all dimensions that has value of 1. E.g. (2,3,1) ==> (2,3)

* Unsqueeze adds a unit dimension to specific axes.

* Permute - Return viewof the input with swapped dimensions


```python
TENSOR =  torch.rand(1,3,4,5,1)
```

### i. Reshape

```python
print(f'Tensor initial shape: {TENSOR.shape}')
print(f'Tensor new shape: {TENSOR.reshape([1,4,5,3]).shape}')
print(f'Tensor new shape: {torch.reshape(TENSOR, [1,4,5,3]).shape}')
```

### ii. Squeeze

```python
# Squeeze by index
print(f'Tensor initial shape: {TENSOR.shape}')
print(f'Tensor new shape: {TENSOR.squeeze(4).shape}')
print(f'Tensor new shape: {torch.squeeze(TENSOR, 4).shape}')
```

```python
# Squeeze all
print(f'Tensor initial shape: {TENSOR.shape}')
print(f'Tensor new shape: {TENSOR.squeeze().shape}')
print(f'Tensor new shape: {torch.squeeze(TENSOR).shape}')
```

### iii. View

It shares the memory of the original input hence a change in the view implies a chnage in the original tensor

```python
# Creating a view
z = TENSOR.view(3,4,5)
print(f'Tensor initial shape: {TENSOR.shape}')
print(f'View new shape: {z.shape}')
```

```python
# Changing a few of the elements of z
z[0,0,0] = 420
print(z[0,0,0]) ## Should print 420
print(TENSOR[0,0,0,0,0]) ## Should print 420 as it is affected by the operation on Z
```

### iv. Stack

* Using `torch.stack()` to stack at a specific diemnsion

* Using `torch.vstack()` to stack at dimension 0.

* Using `torch.hstack()` to stack at dimension 1.

```python
# USing .stack()
TENSOR_2 = torch.rand(3,4,5)
torch.stack([z, TENSOR_2], dim=3).shape
```

```python
# Using .vstack()
torch.vstack((z, TENSOR_2)).shape
```

```python
# Using .hstack()
torch.hstack((z,TENSOR_2)).shape
```

### v. Unsqueeze

```python
print(f'Tensor initial shape: {TENSOR.shape}')
print(f'Tensor new shape: {torch.unsqueeze(TENSOR, 0).shape}')
print(f'Tensor new shape: {TENSOR.unsqueeze(5).shape}')
```

### vi. Permute

```python
# Axis swapping using permute
print(f'Tensor initial shape: {TENSOR.shape}')
print(f'Tensor new shape: {torch.permute(TENSOR, [1, 0, 4, 2, 3]).shape}')
print(f'Tensor new shape: {TENSOR.permute([1, 0, 3, 4, 2]).shape}')
```
