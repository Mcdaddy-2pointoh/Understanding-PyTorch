## g. Tensor Datatypes

* Dtypes number specifies the precison in numerical quantities the number is expressed.

* It is the details in which numbers are expressed


Precision & Storage size:

Float64 >>> Float32 >>> Float16


* Device specifies the processing unit for the tensor, 'cpu' as default and 'cuda' for using gpu as cuda is an interfacing toolkit

* requires_grad specifies the tracking of gradient of change as the tensor based operations progress. Ususally on weight/ bias adjustment

```python
# Float 32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0], dtype=None)
float_32_tensor.dtype
```

```python
# Specifying data type using dtype
int_64_tensor = torch.tensor([3.0, 6.0, 9.0], dtype=torch.int64)
int_64_tensor.dtype
```

```python
# Devices in tensors specifies the processing unit
float_32_tensor = torch.tensor([3.0, 6.0, 9.0], dtype=torch.int64, device='cpu')
float_32_tensor.device
```

```python
# To convert tensor data type
float_64_tensor = float_32_tensor.type(torch.float64)
float_64_tensor
```

```python
float_64_tensor * int_64_tensor
```

### 2. TENSOR Operations


There are two ways to multiply in matices, vectors or tensors:

* Element wise operation as done by `torch.mul(TENSOR, scalar)`


* Matrix multiplication [Dot Product] operation as done by `torch.matmul(TENSOR_A, TENSOR_B)`. This operation can be only done when shape of TENSOR_A is (m,n) and shape of TENSOR_B is (n,o) producing the resulting tensor (TENSOR_R) with a shape (m,o)

```python
# Mat multiplying a vector by its transpose using torch.transpose()
torch.matmul(TENSOR, torch.transpose(TENSOR, 0, 1))
```

```python
# Mat multiplying a vector by its transpose using TENSOR.T
torch.matmul(TENSOR, TENSOR.T)
```

```python
# Understanding the change in TENSOR due to transpose
print(TENSOR)
print(TENSOR.T)
```

```python
# Understanding the change in shape of a transpose
print(TENSOR.shape, TENSOR.T.shape)
```
